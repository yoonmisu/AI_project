```python
# ÏÉòÌîå Ïù¥ÎØ∏ÏßÄ ÌôïÏù∏ Ìï®Ïàò
def show_samples(folder, title):
    files = os.listdir(folder)[:5]  # Ïïû 5Ïû•Îßå ÌôïÏù∏
    plt.figure(figsize=(12, 4))
    for i, file in enumerate(files):
        img_path = os.path.join(folder, file)
        try:
            img = mpimg.imread(img_path)
            plt.subplot(1, 5, i+1)
            plt.imshow(img)
            plt.axis('off')
        except:
            continue
    plt.suptitle(title)
    plt.show()

# ÏÉòÌîå ÌôïÏù∏
show_samples("drive/MyDrive/Jelly_Detergent_dataset/Jelly", "üç¨ Jelly ÏÉòÌîå")
show_samples("drive/MyDrive/Jelly_Detergent_dataset/Detergent", "üßº Detergent ÏÉòÌîå")
```

```python
import os, shutil

src_folder = "dataset/Detergent/yellow laundry detergent pods on white background close up photo"
dst_folder = "dataset/Detergent"

for file in os.listdir(src_folder):
    src_file = os.path.join(src_folder, file)
    dst_file = os.path.join(dst_folder, file)
    # Ïù¥Î¶Ñ Ï§ëÎ≥µ Ïãú _new Î∂ôÏó¨ÏÑú Ïù¥Îèô
    if os.path.exists(dst_file):
        base, ext = os.path.splitext(file)
        dst_file = os.path.join(dst_folder, base + "_new" + ext)
    shutil.move(src_file, dst_file)

shutil.rmtree(src_folder)
print("üßº Detergent Ìè¥Îçî Íµ¨Ï°∞ Ï†ïÎ¶¨ ÏôÑÎ£å!")
```

```python
import os, random
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

def show_random_samples(folder, title, n=10):
    files = os.listdir(folder)
    if len(files) < n:
        n = len(files)  # Ïù¥ÎØ∏ÏßÄÍ∞Ä nÍ∞úÎ≥¥Îã§ Ï†ÅÏúºÎ©¥ ÏûàÎäî ÎßåÌÅºÎßå
    random_files = random.sample(files, n)

    plt.figure(figsize=(12, 4))
    for i, file in enumerate(random_files):
        img_path = os.path.join(folder, file)
        try:
            img = mpimg.imread(img_path)
            plt.subplot(1, n, i+1)
            plt.imshow(img)
            plt.axis('off')
        except:
            continue
    plt.suptitle(title)
    plt.show()

# ÏòàÏãú: Î¨¥ÏûëÏúÑ 5Ïû• ÏÉòÌîå ÌôïÏù∏
show_random_samples("drive/MyDrive/Jelly_Detergent_dataset/Jelly", "üç¨ Jelly random sample")
show_random_samples("drive/MyDrive/Jelly_Detergent_dataset/Detergent", "üßº Detergent random sample")
```

```python
from PIL import Image
import os

input_base = "/content/drive/MyDrive/Jelly_Detergent_dataset"
output_base = "/content/drive/MyDrive/dataset_gray/train"

# Jelly, Detergent Í∞ÅÍ∞Å Î≥ÄÌôò
for category in ["Jelly", "Detergent"]:
    input_folder = os.path.join(input_base, category)
    output_folder = os.path.join(output_base, category)

    os.makedirs(output_folder, exist_ok=True)

    for file in os.listdir(input_folder):
        if file.lower().endswith(('.jpg', '.png', '.jpeg')):
            img = Image.open(os.path.join(input_folder, file))
            gray = img.convert("L")  # L = grayscale
            gray.save(os.path.join(output_folder, file))

print("‚úî ÌùëÎ∞± Ïù¥ÎØ∏ÏßÄ Î≥ÄÌôò ÏôÑÎ£å!")
```

```python
from PIL import Image
import hashlib

def remove_duplicates(folder):
    seen_hashes = set()
    for file in os.listdir(folder):
        path = os.path.join(folder, file)
        try:
            img = Image.open(path).convert("RGB")
            h = hashlib.md5(img.tobytes()).hexdigest()
            if h in seen_hashes:
                os.remove(path)
                print(f"‚ùå Ï§ëÎ≥µ Ïù¥ÎØ∏ÏßÄ ÏÇ≠Ï†ú: {file}")
            else:
                seen_hashes.add(h)
        except:
            continue

remove_duplicates("drive/MyDrive/Jelly_Detergent_dataset/Jelly")
remove_duplicates("drive/MyDrive/Jelly_Detergent_dataset/Detergent")
```
